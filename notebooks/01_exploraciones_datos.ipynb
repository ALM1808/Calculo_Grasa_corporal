```python
# 01.Exploración de Datos (EDA)

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os

# Verificar que la carpeta de datos existe
from pathlib import Path

# Buscar la raíz del proyecto
ROOT_DIR = Path.cwd()
while not (ROOT_DIR / "pyproject.toml").exists() and ROOT_DIR != ROOT_DIR.parent:
    ROOT_DIR = ROOT_DIR.parent

RAW_DATA_PATH = ROOT_DIR / "data" / "raw"

# Verificar archivos disponibles
available_files = os.listdir(RAW_DATA_PATH)
print("Archivos disponibles en data/raw:")
for file in available_files:
    print(f"- {file}")

# Cargar el dataset
df = pd.read_csv(RAW_DATA_PATH / 'gym_members_exercise_tracking.csv')

# Exploración Inicial
print("Primeras Filas del Dataset:")
print(df.head())

print("Información del Dataset:")
df.info()

print("Estadísticas Descriptivas:")
print(df.describe())

# Visualización Inicial
sns.pairplot(df)
plt.show()
```
