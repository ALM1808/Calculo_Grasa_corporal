```python
# 01_Exploracion_datos.ipynb

# ===============================
# üîπ 1. Importaciones
# ===============================
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from pathlib import Path
import os

# ===============================
# üîπ 2. Definir rutas y cargar datos
# ===============================
# Detectar la ra√≠z del proyecto (donde est√° pyproject.toml)
ROOT_DIR = Path.cwd()
while not (ROOT_DIR / "pyproject.toml").exists() and ROOT_DIR != ROOT_DIR.parent:
    ROOT_DIR = ROOT_DIR.parent

RAW_DATA_PATH = ROOT_DIR / "data" / "raw" / "gym_members_exercise_tracking.csv"

if not RAW_DATA_PATH.exists():
    raise FileNotFoundError(f"‚ùå No se encuentra el archivo en: {RAW_DATA_PATH}")

# Cargar el dataset
print("üì• Cargando dataset original...")
df = pd.read_csv(RAW_DATA_PATH)

# ===============================
# üîπ 3. Exploraci√≥n b√°sica
# ===============================
print("\n‚úÖ Primeras filas del dataset:")
print(df.head())

print("\n‚ÑπÔ∏è Informaci√≥n general:")
df.info()

print("\nüìä Estad√≠sticas descriptivas:")
print(df.describe())

# ===============================
# üîπ 4. Visualizaci√≥n general
# ===============================
sns.pairplot(df.select_dtypes(include=['float64', 'int64']))
plt.suptitle("Distribuciones y relaciones", y=1.02)
plt.show()

# ===============================
# üîπ 5. Detecci√≥n de nulos
# ===============================
missing = df.isnull().sum()
print("\n‚ùó Valores nulos por columna:")
print(missing[missing > 0])
```
